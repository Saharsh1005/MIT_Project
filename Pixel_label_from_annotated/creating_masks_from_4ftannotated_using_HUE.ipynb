```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#our original image which should be masked
im=cv2.imread('C:/Users/Saharsh/Desktop/MIT/Dataset/canon4ftannotated/IMG_4727.jpg')
im=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)

#im2 is the image where we will write the labels using masks
'''
0 - background
1 - red/grass
2 - paddy
3 - broad
'''

im2=np.zeros((2592, 3456), np.uint8)
```

```python
plt.imshow(im)
```

```python
#hue saturation value - hue to create masks
hsv = cv2.cvtColor(im, cv2.COLOR_RGB2HSV)
```

<h1>Creating mask for red and putting label in im2</h1>

```python
#Color Range of the image: (hMin = 0 , sMin = 0, vMin = 0), (hMax = 179 , sMax = 255, vMax = 255)

#1 Red  

#giving a hue range to extract all red parts:
low= (0,150,20)
high= (30,255,255)

mask = cv2.inRange(hsv,low,high)
result = cv2.bitwise_and(im, im, mask=mask)

# labeling this extracted parts as 1
im2[mask>0]=1
```

```python
#Displaying extracted red parts

plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```

```python
# Displaying the labelled im2 vs original image
#all red areas are labelled as 1

plt.subplot(1, 2, 1)
plt.imshow(im)
plt.subplot(1, 2, 2)
plt.imshow(im2)
plt.show()
```

<h1>SIMILARLY FOR GREEN AND BLUE </h1>

```python
#2 Green  
low= (31,100,20)
high= (80,255,255)
mask = cv2.inRange(hsv,low,high)
result = cv2.bitwise_and(im, im, mask=mask)
im2[mask>0]=2
```

```python
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```

```python
plt.imshow(im2)
```

```python
#2 Blue  
low= (80,100,20)
high= (179,255,255)
mask = cv2.inRange(hsv,low,high)
result = cv2.bitwise_and(im, im, mask=mask)
im2[mask>0]=3
```

```python
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```
